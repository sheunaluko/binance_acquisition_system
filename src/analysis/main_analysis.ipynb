```r
# set data dir 
data_dir = 'crypto_market_data_1' 

# get the subfolders of days 
data_days = list.files(data_dir, full.names=TRUE)



read_raw_file <- function(fname) { 
    f_str = read_file(fname) 
    f_str 
}

# datatype can be (TRADE,DEPTH,MARK (for futures))
# ex: get_filename("2020_8_01", "ETHUSDT" , "FUTURES" , "TRADE" ) returns: 
# '/home/oluwa/dev/binance_acquisition_system/synced_files/crypto_market_data_1/2020_8_01/ETHUSDT/FUTURES/ETHUSDT_TRADE.json'
get_filename <- function(day,pair,future_or_spot, data_type) { 
    file.path(data_dir,day,pair,future_or_spot,paste0(pair,"_",data_type,".json"))
}

get_file_string <- function(day,pair,future_or_spot,data_type) { 
    read_raw_file(get_filename(day,pair,future_or_spot,data_type))
}

get_trade_table <- function(day,pair,future_or_spot) { 
    fs = read_raw_file(get_filename(day,pair,future_or_spot,"TRADE"))
    chunks = str_split(fs,"\n")[[1]] %>% stri_remove_empty 
    print(paste("building trade table for: ", day, "_" , pair, "/" , future_or_spot))
    result = do.call(rbind,lapply(chunks, function(s) fromJSON(s,simplify=FALSE) )) %>% as.data.frame
    print("done")                        
    result %>% mutate(p=as.numeric(p),q = as.numeric(q),a = as.numeric(a) , t=as_datetime(as.numeric(T)/1000) ) %>% 
                select(t,p,q,m,a,t)
}



analyze_spot_future_spread <- function(day , pair, bin_level) { 
    # takes a day and a pair 
    # and bin_level which currently supports 'minute' n
    # produces a histogram of the dfs (delta futures spot) averaged by (param: 1 minute bins)  
    20 

}
```
