```python
# for each pair we extract the data 

def process_pair_day(day,pair) : 
    (futures_or_spot,pair) = pair.split("_")
    pdata = paf.extract_tbinned_data_from_file(day,pair,futures_or_spot, "TRADE", "T") 
    
    xs = pdata['v_net'] 
    ys = pdata['p_change_percent'] 
    
    # and calculate the best fit line 
    z = np.polyfit(x=xs, y=ys, deg=1)
    p = np.poly1d(z)
    
    raw_pts = [ (xs[i],ys[i]) for i in range(0,len(xs))]
    fit_line = [ (xs[i], p(xs[i])) for i in range(0,len(xs))] 

    fit_slope = p[1]
    
    graph = (hv.Scatter(raw_pts) * hv.Curve(fit_line)).opts(title="{}={}, ({})".format(pair,fit_slope,"pchange vs v_net"))

    return (graph,fit_slope,p,np.mean(pdata['v_tot'])) 
    
    # s1 = hv.Scatter(f_ds,'v_net','p_change_percent',label='futures')
```

```python
data_dic = {} 
for p in pairs : 
    print("processing: {}".format(p))
    data_dic[p] = process_pair_day(day,p) 
    
```
